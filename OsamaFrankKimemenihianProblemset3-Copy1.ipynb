```python
#code was from https://stackoverflow.com/questions/50674011/replace-the-year-in-pandas-datetime-column.
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].mask(data['Yr_Mo_Dy'].dt.year == 2061, data['Yr_Mo_Dy'] + pd.offsets.DateOffset(year=1961))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
us_marriages = pd.read_csv('us-marriages-divorces-1867-2014.csv')
```

```python
plt.plot(us_marriages['Year'], us_marriages['Divorces_per_1000'], label = 'Divorces_per_1000')
plt.plot(us_marriages['Year'], us_marriages['Marriages_per_1000'], label = 'Marriages_per_1000')
plt.legend()
plt.title('us-marriages-divorces-1867-2014')
plt.xlabel('Year')
plt.ylabel('Per 1000 of population')
```

# Question 7

```python
a = us_marriages
b = a.set_index('Year')
b[['Marriages_per_1000', 'Divorces_per_1000']].loc[[1900, 1950, 2000]].plot.bar()
plt.title('us-marriages-divorces-1900,1950,2000')
plt.xlabel('Year')
plt.ylabel('Per 1000 of population')
```
